```python
# loading our database into a Pandas DataFrame object
song_data=pd.read_csv("/Users/swapnakotha/Desktop/CMPE-255/-CMPE255-FinalProject-Team9-main/YearPredictionMSD.txt", names=T_features)
```

```python
X = song_data.iloc[:,1:]  # X contains 90 columns
Y = song_data.iloc[:,0]  # Y contains year column

# data normalizations (scaling down all values to the interval [0,1])
# The years 1922-2011 are scaled down to integers [0,1,2,..., 89] 

X = (X - X.min()) / (X.max() - X.min())  # all values are between 0 and 1 
Y = Y - Y.min()        # The years 1922-2011 are mapped to 0-89

# Training _data set
X_train_data = X[0:463715]
y_train_data = Y[0:463715]

# Validation _data set
X_test_data = X[463715:]
y_test_data = Y[463715:]
```

```python
X_train_data = X[0:463715].values
y_train_data = Y[0:463715].values

# Validation _data set
X_test_data = X[463715:].values
y_test_data = Y[463715:].values
```

```python
#using LinearRegression
classifier =LinearRegression().fit(X_train_data, y_train_data)
```

```python
y_train_predicted_labels =classifier.predict(X_train_data)
print("Year labels predicted by the classifier on test data",y_train_predicted_labels)
print("Actual labels of test data",y_test_data)
errors_train = mean_squared_error(y_train_data,y_train_predicted_labels,squared=False)
print("RMSE for test data",errors_train)
```

```python
#plot the actual of predicted labels of train data
X=y_train_data[:100]
Y=y_train_predicted_labels[:100]
fig = plt.subplots(figsize=(10, 6), facecolor='white', edgecolor='white')
plt.title('Year labels for 100 songs')
plt.xlabel('Actual year labesl')
plt.ylabel('Predicted year labels')
plt.scatter(X,Y, s = np.pi * (8)**2, c = Y , alpha=0.7)
plt.show()
```

```python
#using LinearRegression
classifier =LinearRegression().fit(X_train_data, y_train_data)
y_test_predicted_labels =classifier.predict(X_test_data)
print("Year labels predicted by the classifier on test data",y_test_predicted_labels)
print("Actual labels of test data",y_test_data)
errors_test = mean_squared_error(y_test_data,y_test_predicted_labels,squared=False)
print("RMSE for test data",errors_test)
```
